```python
import tensorflow as tf
```

```python
IMG_SIZE= 224
BATCH_SIZE= 11
LEARNING_RATE= 0.0005

def load_image( filename: str, label: int):
    file = tf.io.read_file(filename)
    
    img = tf.image.decode_png(file, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE], )
    return  img, label
```

```python
from sklearn.model_selection import train_test_split

classes = dict(Normal= 0, RL=1, LR= 2)
class_names = list(classes.keys())

X_train, X_valid, y_train, y_valid = train_test_split( Bicaval_df[['Age', 'Image']], Dir.map(classes), random_state=14)


```

### For Direction

```python
y_train
```

```python
train_ds = tf.data.Dataset.from_tensor_slices((X_train['Image'], y_train)).map(load_image).batch(BATCH_SIZE)
validation_ds = tf.data.Dataset.from_tensor_slices((X_valid['Image'], y_valid)).map(load_image).batch(BATCH_SIZE)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))
for images, labels in train_ds.take(1):
    labels = labels.numpy()    
    for i in range(10):
        ax = plt.subplot(2, 5, i + 1)
        plt.imshow(images[i].numpy().astype(np.uint8))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Conv2D, Dense, MaxPooling2D, Dropout



model= Sequential([  
    Conv2D(32, (3,3), activation= "relu", input_shape= (224,224, 3)),
    MaxPooling2D(),
    Dropout(LEARNING_RATE),
    Conv2D(64, (3,3), activation= "relu"),
    MaxPooling2D(),
    Dropout(LEARNING_RATE),
    Conv2D(32, (3,3), activation= "relu"), MaxPooling2D(), Dropout(LEARNING_RATE),
    Conv2D(32, (3,3), activation= "relu"),
    MaxPooling2D(),
    #Dropout(LEARNING_RATE),
    Conv2D(24, (3,3), activation= "relu"),
    MaxPooling2D(),
    Dropout(LEARNING_RATE),
    Flatten(),
    Dense(100, activation= "relu"),
    Dense(10, activation= "relu"),
    Dense(3, activation= "softmax") #Because the labels are 0,1,2
])
model.summary() 
```

```python
model.compile(
optimizer= "adam",
loss= "sparse_categorical_crossentropy",
metrics= ["accuracy"]
)
```

```python
model.fit(train_ds, batch_size= 19,  epochs= 23)
```

```python
predictions= model.predict(validation_ds)
predicted_class = np.argmax(predictions, axis=1)

for i, (pred, actual) in enumerate(zip(predicted_class, labels)):
        ax = plt.subplot (2, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        if pred == actual:
            plt.title(class_names[labels[i]], color="green", size=9)
        else:
            plt.title(f"{class_names[predicted_class[i]]}\n" + f"(Actual: {class_names[labels[i]]})", color="red", size=9)
            plt.axis("off")
```

```python
loss, accuracy = model.evaluate(validation_ds, verbose=0)
print(f"Accuracy:  {accuracy:.5f}",
          f"Loss:      {loss:.5f}\n", sep="\n")
```




We shall try the [U-Net architecture](https://www.sciencedirect.com/science/article/pii/S2352914821000915) on the data, because of its repute in heart image segmentation. 

```python
from tensorflow.keras.layers import UpSampling2D

model1= Sequential([  
    Conv2D(64, (3,3), activation= "relu", input_shape= (IMG_SIZE, IMG_SIZE, 3)),
    MaxPooling2D(),
    Conv2D(128, (3,3), activation= "relu"),MaxPooling2D(),
    Conv2D(32, (3,3), activation= "relu"),MaxPooling2D(),
    Dropout(LEARNING_RATE),
    Conv2D(24, (3,3), activation= "relu"),
    Dropout(LEARNING_RATE),
    
    
    UpSampling2D(size=(2, 2), data_format="channels_last", interpolation='gaussian'),
    UpSampling2D(size=(2, 2), data_format="channels_last", interpolation='gaussian'),
    #UpSampling2D(size=(3, 3), data_format="channels_last", interpolation='gaussian'),
    Conv2D(32, (3,3), activation= "relu"),MaxPooling2D(),
    Dropout(LEARNING_RATE),
    
    UpSampling2D(size=(2, 2), data_format="channels_last", interpolation='gaussian'),
    Conv2D(24, (3,3), activation= "relu"),MaxPooling2D(),
    UpSampling2D(size=(2, 2), data_format="channels_last", interpolation='gaussian'),
    Conv2D(24, (3,3), activation= "relu"),MaxPooling2D(),
    UpSampling2D(size=(2, 2), data_format="channels_last", interpolation='gaussian'),
    Conv2D(24, (3,3), activation= "relu"),MaxPooling2D(),
   
   
    Dropout(LEARNING_RATE),
    Conv2D(24, (3,3), activation= "relu"),MaxPooling2D(),
    Dropout(LEARNING_RATE),
    
    Conv2D(24, (3,3), activation= "relu"),MaxPooling2D(),
    Dropout(LEARNING_RATE),
    Flatten(),

    Dense(100, activation= 'relu'),
    Dense(3, activation= 'softmax')

])

model1.summary()
```

```python
model1.compile(
optimizer= "adam",
loss= "sparse_categorical_crossentropy",
metrics= ["accuracy"]
)
```

```python
model1.fit(train_ds, batch_size= BATCH_SIZE,  epochs= 23)
```

```python
predictions1= model1.predict(validation_ds)
predicted_class1 = np.argmax(predictions1, axis=1)

for i, (pred, actual) in enumerate(zip(predicted_class1, labels)):
        ax = plt.subplot (2, 5, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        if pred == actual:
            plt.title(class_names[labels[i]], color="green", size=9)
        else:
            plt.title(f"{class_names[predicted_class1[i]]}\n" + f"(Actual: {class_names[labels[i]]})", color="red", size=9)
            plt.axis("off")
```

```python
loss1, accuracy1 = model1.evaluate(validation_ds, verbose=0)
print(f"Accuracy:  {accuracy:.5f}",
          f"Loss:      {loss:.5f}\n", sep="\n")
   
```
